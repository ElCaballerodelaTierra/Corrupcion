# Visualización de la red por medio de mathplotlib

```python
#Importamos librerías
import networkx as nx
import matplotlib.pyplot as plt


#Creamos la red
 
#Una red vacía
G = nx.Graph()

#Añadimos 100 nodos
G.add_nodes_from(range(1, 101))  # Añade los nodos del 1 al 100

#Visualizar la red

#Posición de los nodos, indica al algoritmo que acomoda los nodos en el espacio
    #La constante k aumenta la repulsión entre nodos
ps = nx.spring_layout(G, k=0.01)
#Indica el tamaño de los nodos
nz = 1000
#Indica el tamaño de la fuente de las etiquetas para los nodos
fz = 16
#Indica el grosor de las aristas
w = 2
#Tipo de tipografía
ff = 'Latin Modern Roman'
#Color de los nodos, azul lila ligero
nc = '#9FA3D5'
#Color de los enlaces, azul morado oscuro
ec = '#3B4285'


#Se dibuja la red
plt.figure(figsize=(10, 10))  # Ajustar el tamaño de la figura
nx.draw(G, pos=ps, with_labels=False, font_size=fz ,node_size=nz ,node_color=nc, edge_color= ec, width = w, font_weight='normal', font_family = ff)
plt.show()
```

```python

# Crear una red Barabási-Albert
n = 100  # Aumentar el número de nodos
m = 1    # Cada nodo nuevo se conecta a 3 nodos existentes
BA_graph = nx.barabasi_albert_graph(n, m)


#Indica el tamaño de los nodos
nz = 400
#Indica el tamaño de la fuente de las etiquetas para los nodos
fz = 16
#Indica el grosor de las aristas
w = 2
#Color de los nodos, azul lila ligero
nc = '#9FA3D5'
#Color de los enlaces, azul morado oscuro
ec = '#3B4285'


#Se dibuja la red

# Algoritmo de orden de nodos
ps = nx.kamada_kawai_layout(BA_graph)
# Ajustar el tamaño de la figura
plt.figure(figsize=(15, 15))  

# Dibujar nodos, aristas y etiquetas
nx.draw(BA_graph, pos=ps, with_labels=False, font_size=fz ,node_size=nz ,node_color=nc, edge_color= ec, width = w, font_weight='normal', font_family = ff)
plt.show()
```
