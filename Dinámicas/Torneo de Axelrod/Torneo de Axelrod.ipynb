```python
import axelrod as axl

# Definir una nueva matriz de pago
new_payoff_matrix = {
    ('C', 'C'): (3, 3),
    ('C', 'D'): (0, 5),
    ('D', 'C'): (5, 0),
    ('D', 'D'): (1, 1)
}

# Crear un nuevo juego con la nueva matriz de pago
new_game = axl.Game(r=new_payoff_matrix[('C', 'C')][0],
                    s=new_payoff_matrix[('C', 'D')][0],
                    t=new_payoff_matrix[('D', 'C')][0],
                    p=new_payoff_matrix[('D', 'D')][0])

# Verificar la nueva matriz de pago
print(new_game.RPST())
# Definir una nueva matriz de pago con 4 estrategias
new_payoff_matrix_4_strategies = {
    ('A', 'A'): (3, 3), ('A', 'B'): (0, 5), ('A', 'C'): (2, 2), ('A', 'D'): (1, 4),
    ('B', 'A'): (5, 0), ('B', 'B'): (1, 1), ('B', 'C'): (4, 3), ('B', 'D'): (2, 2),
    ('C', 'A'): (2, 2), ('C', 'B'): (3, 4), ('C', 'C'): (3, 3), ('C', 'D'): (0, 5),
    ('D', 'A'): (4, 1), ('D', 'B'): (2, 2), ('D', 'C'): (5, 0), ('D', 'D'): (1, 1)
}

# Crear un nuevo juego con la nueva matriz de pago de 4 estrategias
new_game_4_strategies = axl.Game(
    r=new_payoff_matrix_4_strategies[('A', 'A')][0],
    s=new_payoff_matrix_4_strategies[('A', 'B')][0],
    t=new_payoff_matrix_4_strategies[('B', 'A')][0],
    p=new_payoff_matrix_4_strategies[('B', 'B')][0]
)

# Verificar la nueva matriz de pago de 4 estrategias
print(new_game_4_strategies.RPST())
```
