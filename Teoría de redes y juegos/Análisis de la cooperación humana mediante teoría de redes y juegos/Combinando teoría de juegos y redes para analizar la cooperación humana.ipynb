## Combining evolutionaty game theory with network theory to analyze human cooperation patterns, Scatà et al. (2016)

Imagina una comunidad de personas conectadas entre sí, como una red social. Cada persona puede elegir ser **cooperativa** o **defectar** (ser egoísta).  Piensa en esto como si estuvieran jugando un juego donde pueden ayudarse mutuamente o intentar sacar ventaja.

El modelo de los investigadores estudia cómo la cooperación se extiende en esta red a lo largo del tiempo.  Para hacer esto, usan ideas de la **Teoría de Juegos Evolutiva**, que es como estudiar juegos, pero pensando en cómo las estrategias (cooperar o defectar) cambian y se propagan en una población con el tiempo.

**Elementos Clave del Modelo:**

1. **Red Social:**  Las personas están conectadas en una red. Algunas personas tienen más conexiones que otras (como en las redes sociales reales). Usan un tipo específico de red llamada "libre de escala", que imita las redes del mundo real.

2. **Juegos Sociales:**  Usan dos juegos clásicos para representar situaciones donde la cooperación es un dilema:
    * **Dilema del Prisionero:**  En este juego, si ambos cooperan, ambos ganan algo. Si ambos defectan, ambos pierden un poco. Pero si uno coopera y el otro defecta, el que defecta gana mucho y el que coopera pierde mucho.  Aquí, la tentación es siempre defectar, aunque la cooperación mutua sería mejor para ambos.
    * **Juego de la Montaña de Nieve (Snowdrift Game):** Similar al dilema del prisionero, pero con una pequeña diferencia.  Aquí, si ambos cooperan, ambos ganan algo bueno. Si ambos defectan, ambos pierden mucho.  Pero si uno coopera y el otro defecta, el que coopera pierde un poco menos que si ambos defectaran, y el que defecta gana mucho.  En este juego, la cooperación puede ser más estable porque si nadie coopera, las cosas van muy mal para todos.

3. **Masa Crítica:**  Los investigadores introducen la idea de "Masa Crítica".  Esto es como la **cantidad mínima de personas cooperativas que se necesitan al principio para que la cooperación se propague** a toda la red.  Si tienes suficiente "semilla" de cooperadores, la cooperación puede empezar a crecer.

4. **Centralidad:**  No todas las personas en la red son igual de importantes. Algunas son más "centrales", es decir, tienen más conexiones y pueden influir más en otros.  Los investigadores exploran si es mejor que la "Masa Crítica" de cooperadores iniciales esté formada por personas **centrales** (influyentes) o **periféricas** (menos influyentes) en la red.

5. **Homofilia:**  Este es un concepto sociológico que significa "los iguales se juntan".  En términos simples, **las personas tienden a conectarse y cooperar más con personas que son similares a ellas**.  Los investigadores incorporan la homofilia al modelo, asumiendo que las personas son más propensas a copiar la estrategia (cooperar o defectar) de alguien similar a ellas.

**Cómo Funciona el Modelo Paso a Paso:**

1. **Crear la Red:** Se crea una red social "libre de escala".
2. **Elegir la Masa Crítica:** Se decide cuántas personas serán cooperativas al inicio (la Masa Crítica).  Luego, se eligen *quiénes* serán estos cooperadores iniciales:
    * **Nodos Centrales:**  Se eligen las personas más influyentes (centrales) de la red.
    * **Nodos Periféricos:** Se eligen las personas menos influyentes (periféricas) de la red.
    * **Nodos Aleatorios:** Se eligen personas al azar.
3. **Asignar Estrategias Iniciales:**  A la Masa Crítica se le asigna la estrategia de "cooperar". Al resto de la gente se le asigna la estrategia de "defectar".
4. **Jugar el Juego:**  Durante varias rondas:
    * Cada persona juega el juego (Dilema del Prisionero o Juego de la Montaña de Nieve) con todos sus vecinos en la red.
    * Cada persona calcula su "ganancia" (payoff) total de estas interacciones.
5. **Actualizar Estrategias:**  Después de cada ronda, cada persona tiene la oportunidad de cambiar su estrategia.  La probabilidad de cambiar a la estrategia de un vecino depende de dos cosas:
    * **Diferencia de Ganancia:** Si un vecino tiene una ganancia mucho mayor, es más probable que la persona cambie a la estrategia de ese vecino.
    * **Homofilia:**  Si un vecino es "similar" (homófilo), es más probable que la persona copie su estrategia, incluso si la diferencia de ganancia no es tan grande.  (En el modelo, la "similitud" se representa con un valor numérico).
6. **Repetir:** Se repiten los pasos 4 y 5 durante muchas rondas para ver cómo evoluciona la cooperación en la red.

**Resultados Principales:**

Los investigadores encontraron que:

* **La homofilia ayuda a la cooperación a extenderse más rápido.** Cuando las personas tienden a imitar a quienes son similares, la cooperación puede consolidarse en grupos de personas similares y luego extenderse más fácilmente.
* **La centralidad de la Masa Crítica importa.**  Iniciar la cooperación con personas **centrales** generalmente lleva a una propagación más rápida de la cooperación. Sin embargo, iniciar con personas **periféricas** (nodos de borde) con alta homofilia también puede ser efectivo, ya que estos nodos pueden formar grupos cooperativos cohesivos que, aunque menos influyentes globalmente, pueden impulsar la cooperación localmente y eventualmente a la red.
* **El tipo de juego influye.**  La cooperación es más difícil de lograr en el Dilema del Prisionero que en el Juego de la Montaña de Nieve, como era de esperar, ya que el Dilema del Prisionero es un escenario más estrictamente no cooperativo.

**En Resumen:**

El modelo muestra cómo la combinación de **Masa Crítica inicial, centralidad en la red y homofilia** juega un papel crucial en la evolución de la cooperación humana en redes sociales.  Entender estos factores puede ayudarnos a diseñar estrategias para promover la cooperación en diferentes contextos sociales.

**Explicación del código:**

* **`initialize_strategies(...)`:**  Asigna estrategias iniciales. Los nodos de la Masa Crítica inicial se marcan como cooperadores ('C'), y el resto como defectores ('D').
* **`calculate_payoff(...)`:**  Define las matrices de recompensa para el Dilema del Prisionero (PDG) y el Juego de la Montaña de Nieve (SG) y calcula la recompensa para un jugador en una interacción.
* **`update_strategy(...)`:** Implementa la regla de actualización de estrategias de Fermi, **simplificada para incluir un factor de homofilia**.  La probabilidad de copiar la estrategia de un vecino se reduce si el nodo y el vecino tienen la misma estrategia (simulando homofilia).  La función de Fermi controla la probabilidad de cambio basada en la diferencia de recompensas y el ruido `K`.
* **`run_simulation(...)`:**  Ejecuta la simulación completa:
    * Inicializa estrategias y registra la historia de la cooperación.
    * En cada ronda:
        * Calcula las recompensas para todos los nodos jugando con sus vecinos.
        * Actualiza las estrategias de todos los nodos usando `update_strategy`.
        * Calcula y guarda la fracción de cooperadores.
    * Retorna la historia de la fracción de cooperación a lo largo de las rondas.
* **`if __name__ == "__main__":`:**  Esta sección ejecuta la simulación principal cuando ejecutas el script directamente:

* **Cambia `cm_placement_type`**  en el bucle `for` o directamente en la sección `if __name__ == "__main__":` para probar diferentes tipos de colocación de la Masa Crítica (central, periférica, aleatoria).

Este modelo es una versión simplificada del descrito en el artículo, pero captura las dinámicas fundamentales y te permite explorar el impacto de la Masa Crítica, la centralidad y la homofilia en la evolución de la cooperación en redes sociales. 

````python
import networkx as nx
import numpy as np
import random
import matplotlib.pyplot as plt

def create_scale_free_network(n, m=2):
    """Crea una red libre de escala usando el modelo Barabási-Albert."""
    return nx.barabasi_albert_graph(n, m)

def initialize_strategies(network, initial_cooperators_nodes, strategy_cooperate='C', strategy_defect='D'):
    """Inicializa las estrategias de los nodos.

    Args:
        network: Objeto de red de NetworkX.
        initial_cooperators_nodes: Lista de nodos iniciales cooperadores.
        strategy_cooperate: Símbolo para la estrategia de cooperación.
        strategy_defect: Símbolo para la estrategia de defección.

    Returns:
        Diccionario con las estrategias iniciales de cada nodo.
    """
    strategies = {}
    for node in network.nodes():
        if node in initial_cooperators_nodes:
            strategies[node] = strategy_cooperate
        else:
            strategies[node] = strategy_defect
    return strategies

def calculate_payoff(player_strategy, neighbor_strategy, game_type="PDG"):
    """Calcula la recompensa (payoff) para un jugador en una interacción.

    Args:
        player_strategy: Estrategia del jugador ('C' o 'D').
        neighbor_strategy: Estrategia del vecino ('C' o 'D').
        game_type: Tipo de juego ("PDG" para Dilema del Prisionero, "SG" para Snowdrift Game).

    Returns:
        Recompensa (payoff) para el jugador.
    """
    # Matrices de recompensa (Payoff Matrices)
    if game_type == "PDG":
        payoff_matrix = {
            ('C', 'C'): (3, 3),  # (Recompensa jugador, Recompensa vecino)
            ('C', 'D'): (0, 5),
            ('D', 'C'): (5, 0),
            ('D', 'D'): (1, 1)
        }
    elif game_type == "SG": # Snowdrift Game
        payoff_matrix = {
            ('C', 'C'): (3, 3),
            ('C', 'D'): (1, 5),
            ('D', 'C'): (5, 1),
            ('D', 'D'): (0, 0)
        }
    else:
        raise ValueError("Tipo de juego no válido. Debe ser 'PDG' o 'SG'.")

    return payoff_matrix[(player_strategy, neighbor_strategy)][0]


def update_strategy(network, node, strategies, payoffs, homophily_level, K=0.1, game_type="PDG"):
    """Actualiza la estrategia de un nodo basándose en la regla de Fermi.

    Args:
        network: Objeto de red de NetworkX.
        node: Nodo cuya estrategia se va a actualizar.
        strategies: Diccionario de estrategias actuales.
        payoffs: Diccionario de recompensas (payoffs) de la ronda anterior.
        homophily_level: Nivel de homofilia (valor entre 0 y 1, 0 sin homofilia, 1 homofilia total).
        K: Temperatura/ruido en la función de Fermi (controla la aleatoriedad de la imitación).
        game_type: Tipo de juego ("PDG" o "SG").

    Returns:
        Nueva estrategia para el nodo.
    """
    current_strategy = strategies[node]
    current_payoff = payoffs[node]

    neighbor = random.choice(list(network.neighbors(node))) # Elige un vecino al azar
    neighbor_strategy = strategies[neighbor]
    neighbor_payoff = payoffs[neighbor]

    if current_strategy == neighbor_strategy: # Simplificación de la homofilia: Si son iguales, menos probable cambiar
        homophily_factor = 1 - homophily_level # Reducir la prob. de cambio si hay homofilia
    else:
        homophily_factor = 1 # Sin efecto de homofilia si son diferentes

    # Función de Fermi modificada con homofilia (simplificada)
    probability_to_adopt = 1 / (1 + np.exp((current_payoff - neighbor_payoff) / K) ) * homophily_factor


    if random.random() < probability_to_adopt:
        return neighbor_strategy # Adopta la estrategia del vecino
    else:
        return current_strategy # Mantiene la estrategia actual


def run_simulation(network, initial_cm_size, cm_placement_type, homophily_level, K, game_type, rounds=200):
    """Ejecuta la simulación del juego evolutivo.

    Args:
        network: Objeto de red de NetworkX.
        initial_cm_size: Tamaño de la Masa Crítica inicial (número de cooperadores iniciales).
        cm_placement_type: Tipo de colocación de la Masa Crítica ('central', 'edge', 'random').
        homophily_level: Nivel de homofilia.
        K: Temperatura/ruido en la función de Fermi.
        game_type: Tipo de juego ("PDG" o "SG").
        rounds: Número de rondas de simulación.

    Returns:
        Lista con la fracción de cooperadores en cada ronda.
    """

    n_nodes = len(network.nodes())
    centrality = nx.eigenvector_centrality(network) # Calcular centralidad

    # Seleccionar nodos de la Masa Crítica según el tipo de colocación
    if cm_placement_type == 'central':
        cm_nodes = sorted(centrality, key=centrality.get, reverse=True)[:initial_cm_size] # Nodos más centrales
    elif cm_placement_type == 'edge':
        cm_nodes = sorted(centrality, key=centrality.get)[:initial_cm_size] # Nodos menos centrales (de borde)
    elif cm_placement_type == 'random':
        cm_nodes = random.sample(list(network.nodes()), initial_cm_size) # Nodos aleatorios
    else:
        raise ValueError("Tipo de colocación de Masa Crítica no válido. Debe ser 'central', 'edge' o 'random'.")


    strategies = initialize_strategies(network, cm_nodes)
    cooperation_fraction_history = []

    for _ in range(rounds):
        payoffs = {node: 0 for node in network.nodes()} # Reiniciar payoffs en cada ronda

        # Jugar el juego y calcular payoffs
        for node in network.nodes():
            for neighbor in network.neighbors(node):
                payoffs[node] += calculate_payoff(strategies[node], strategies[neighbor], game_type)

        # Actualizar estrategias para la siguiente ronda
        next_strategies = {}
        for node in network.nodes():
            next_strategies[node] = update_strategy(network, node, strategies, payoffs, homophily_level, K, game_type)
        strategies = next_strategies # Actualizar estrategias para la siguiente ronda

        cooperation_count = sum(1 for strategy in strategies.values() if strategy == 'C')
        cooperation_fraction = cooperation_count / n_nodes
        cooperation_fraction_history.append(cooperation_fraction)

    return cooperation_fraction_history


if __name__ == "__main__":
    n_nodes = 100
    initial_cm_size = 10
    homophily_level = 0.5  # Nivel de homofilia (entre 0 y 1)
    K = 0.1 # Ruido en la función de Fermi
    game_type = "PDG" # "PDG" (Dilema del Prisionero) o "SG" (Snowdrift Game)
    rounds = 200

    cm_placements = ['central', 'edge', 'random']
    placement_results = {}

    for cm_placement_type in cm_placements:
        network = create_scale_free_network(n_nodes)
        cooperation_history = run_simulation(network, initial_cm_size, cm_placement_type, homophily_level, K, game_type, rounds)
        placement_results[cm_placement_type] = cooperation_history

    # Graficar resultados
    plt.figure(figsize=(10, 6))
    for placement_type, history in placement_results.items():
        plt.plot(history, label=f'CM Placement: {placement_type}')

    plt.xlabel("Ronda")
    plt.ylabel("Fracción de Cooperadores")
    plt.title(f"Evolución de la Cooperación ({game_type}, Homofilia={homophily_level}, K={K})")
    plt.legend()
    plt.grid(True)
    plt.show()
````
