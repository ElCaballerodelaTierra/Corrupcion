## Combining evolutionaty game theory with network theory to analyze human cooperation patterns, Scatà et al. (2016)

Imagina una comunidad de personas conectadas entre sí, como una red social. Cada persona puede elegir ser **cooperativa** o **defectar** (ser egoísta).  Piensa en esto como si estuvieran jugando un juego donde pueden ayudarse mutuamente o intentar sacar ventaja.

El modelo de los investigadores estudia cómo la cooperación se extiende en esta red a lo largo del tiempo.  Para hacer esto, usan ideas de la **Teoría de Juegos Evolutiva**

**Elementos Clave del Modelo:**

1. **Red Social:**  Las personas están conectadas en una red. Algunas personas tienen más conexiones que otras (como en las redes sociales reales). Usan un tipo específico de red llamada "libre de escala", que imita las redes del mundo real.

2. **Juegos Sociales:**  Usan dos juegos clásicos para representar situaciones donde la cooperación es un dilema:
    * **Dilema del Prisionero:**  En este juego, si ambos cooperan, ambos ganan algo. Si ambos defectan, ambos pierden un poco. Pero si uno coopera y el otro defecta, el que defecta gana mucho y el que coopera pierde mucho.  Aquí, la tentación es siempre defectar, aunque la cooperación mutua sería mejor para ambos.
    * **Juego de la Montaña de Nieve (Snowdrift Game):** Similar al dilema del prisionero, pero con una pequeña diferencia.  Aquí, si ambos cooperan, ambos ganan algo bueno. Si ambos defectan, ambos pierden mucho.  Pero si uno coopera y el otro defecta, el que coopera pierde un poco menos que si ambos defectaran, y el que defecta gana mucho.  En este juego, la cooperación puede ser más estable porque si nadie coopera, las cosas van muy mal para todos.

3. **Masa Crítica:**  Los investigadores introducen la idea de "Masa Crítica".  Esto es como la **cantidad mínima de personas cooperativas que se necesitan al principio para que la cooperación se propague** a toda la red.  Si tienes suficiente "semilla" de cooperadores, la cooperación puede empezar a crecer.

4. **Centralidad:**  No todas las personas en la red son igual de importantes. Algunas son más "centrales", es decir, tienen más conexiones y pueden influir más en otros.  Los investigadores exploran si es mejor que la "Masa Crítica" de cooperadores iniciales esté formada por personas **centrales** (influyentes) o **periféricas** (menos influyentes) en la red.

5. **Homofilia:**  Este es un concepto sociológico que significa "los iguales se juntan".  En términos simples, **las personas tienden a conectarse y cooperar más con personas que son similares a ellas**.  Los investigadores incorporan la homofilia al modelo, asumiendo que las personas son más propensas a copiar la estrategia (cooperar o defectar) de alguien similar a ellas.

**Cómo Funciona el Modelo Paso a Paso:**

1. **Crear la Red:** Se crea una red social "libre de escala".
2. **Elegir la Masa Crítica:** Se decide cuántas personas serán cooperativas al inicio (la Masa Crítica).  Luego, se eligen *quiénes* serán estos cooperadores iniciales:
    * **Nodos Centrales:**  Se eligen las personas más influyentes (centrales) de la red.
    * **Nodos Periféricos:** Se eligen las personas menos influyentes (periféricas) de la red.
    * **Nodos Aleatorios:** Se eligen personas al azar.
3. **Asignar Estrategias Iniciales:**  A la Masa Crítica se le asigna la estrategia de "cooperar". Al resto de la gente se le asigna la estrategia de "defectar".
4. **Jugar el Juego:**  Durante varias rondas:
    * Cada persona juega el juego (Dilema del Prisionero o Juego de la Montaña de Nieve) con todos sus vecinos en la red.
    * Cada persona calcula su "ganancia" (payoff) total de estas interacciones.
5. **Actualizar Estrategias:**  Después de cada ronda, cada persona tiene la oportunidad de cambiar su estrategia.  La probabilidad de cambiar a la estrategia de un vecino depende de dos cosas:
    * **Diferencia de Ganancia:** Si un vecino tiene una ganancia mucho mayor, es más probable que la persona cambie a la estrategia de ese vecino.
    * **Homofilia:**  Si un vecino es "similar" (homófilo), es más probable que la persona copie su estrategia, incluso si la diferencia de ganancia no es tan grande.  (En el modelo, la "similitud" se representa con un valor numérico).
6. **Repetir:** Se repiten los pasos 4 y 5 durante muchas rondas para ver cómo evoluciona la cooperación en la red.

**Resultados Principales:**

Los investigadores encontraron que:

* **La homofilia ayuda a la cooperación a extenderse más rápido.** Cuando las personas tienden a imitar a quienes son similares, la cooperación puede consolidarse en grupos de personas similares y luego extenderse más fácilmente.
* **La centralidad de la Masa Crítica importa.**  Iniciar la cooperación con personas **centrales** generalmente lleva a una propagación más rápida de la cooperación. Sin embargo, iniciar con personas **periféricas** (nodos de borde) con alta homofilia también puede ser efectivo, ya que estos nodos pueden formar grupos cooperativos cohesivos que, aunque menos influyentes globalmente, pueden impulsar la cooperación localmente y eventualmente a la red.
* **El tipo de juego influye.**  La cooperación es más difícil de lograr en el Dilema del Prisionero que en el Juego de la Montaña de Nieve, como era de esperar, ya que el Dilema del Prisionero es un escenario más estrictamente no cooperativo.

**En Resumen:**

El modelo muestra cómo la combinación de **Masa Crítica inicial, centralidad en la red y homofilia** juega un papel crucial en la evolución de la cooperación humana en redes sociales.  Entender estos factores puede ayudarnos a diseñar estrategias para promover la cooperación en diferentes contextos sociales.

````python
import networkx as nx
import numpy as np
import random
from IPython.display import SVG, display
import matplotlib.pyplot as plt
import scienceplots  # Importar la librería scienceplots

def crear_red_libre_escala(n, m=1):
    """Crea una red libre de escala utilizando el modelo Barabási-Albert."""
    return nx.barabasi_albert_graph(n, m)

def inicializar_estrategias(red, nodos_cooperadores_iniciales, estrategia_cooperar='C', estrategia_defectar='D'):
    """Inicializa las estrategias de los nodos de la red."""
    estrategias = {}
    for nodo in red.nodes():
        if nodo in nodos_cooperadores_iniciales:
            estrategias[nodo] = estrategia_cooperar
        else:
            estrategias[nodo] = estrategia_defectar
    return estrategias

def calcular_recompensa(estrategia_jugador, estrategia_vecino, tipo_juego="PDG"):
    """Calcula la recompensa (payoff) para un jugador en una interacción con un vecino."""
    # Matrices de recompensa (Payoff Matrices)
    if tipo_juego == "PDG":
        matriz_recompensa = {
            ('C', 'C'): (3, 3),  # (Recompensa jugador, Recompensa vecino)
            ('C', 'D'): (0, 5),
            ('D', 'C'): (5, 0),
            ('D', 'D'): (1, 1)
        }
    elif tipo_juego == "SG": # Juego de la Montaña de Nieve (Snowdrift Game)
        matriz_recompensa = {
            ('C', 'C'): (3, 3),
            ('C', 'D'): (1, 5),
            ('D', 'C'): (5, 1),
            ('D', 'D'): (0, 0)
        }
    else:
        raise ValueError("Tipo de juego no válido. Debe ser 'PDG' o 'SG'.")

    return matriz_recompensa[(estrategia_jugador, estrategia_vecino)][0]


def actualizar_estrategia(red, nodo, estrategias, recompensas, nivel_homofilia, K=0.1, tipo_juego="PDG"):
    """Actualiza la estrategia de un nodo utilizando la regla de Fermi, con influencia de la homofilia."""
    estrategia_actual = estrategias[nodo]
    recompensa_actual = recompensas[nodo]

    vecino = random.choice(list(red.neighbors(nodo))) # Elige un vecino al azar
    estrategia_vecino = estrategias[vecino]
    recompensa_vecino = recompensas[vecino]

    if estrategia_actual == estrategia_vecino: # Simplificación de la homofilia: Menos probable cambiar si las estrategias son iguales
        factor_homofilia = 1 - nivel_homofilia # Reduce la probabilidad de cambio si hay homofilia
    else:
        factor_homofilia = 1 # Sin efecto de homofilia si las estrategias son diferentes
        
    # Función de Fermi modificada con un factor de homofilia simplificado
    try:
        probabilidad_adoptar = (1 / (1 + np.exp((recompensa_actual - recompensa_vecino) / K))) * factor_homofilia
    except OverflowError:
        probabilidad_adoptar = 0 if (recompensa_actual - recompensa_vecino) > 0 else 1

    if random.random() < probabilidad_adoptar:
        return estrategia_vecino # Adopta la estrategia del vecino
    else:
        return estrategia_actual # Mantiene la estrategia actual


def ejecutar_simulacion(red, tamaño_masa_critica_inicial, tipo_colocacion_masa_critica, nivel_homofilia, K, tipo_juego, rondas=200):
    """Ejecuta la simulación completa del juego evolutivo en la red y retorna la red y estrategias finales."""
    num_nodos = len(red.nodes())
    centralidad = nx.eigenvector_centrality(red, max_iter=1000) # Calcular la centralidad de vector propio de los nodos

    # Seleccionar los nodos de la Masa Crítica inicial basándose en el tipo de colocación
    if tipo_colocacion_masa_critica == 'central':
        nodos_masa_critica = sorted(centralidad, key=centralidad.get, reverse=True)[:tamaño_masa_critica_inicial] # Nodos más centrales
    elif tipo_colocacion_masa_critica == 'edge':
        nodos_masa_critica = sorted(centralidad, key=centralidad.get)[:tamaño_masa_critica_inicial] # Nodos menos centrales (de borde)
    elif tipo_colocacion_masa_critica == 'random':
        nodos_masa_critica = random.sample(list(red.nodes()), tamaño_masa_critica_inicial) # Nodos aleatorios
    else:
        raise ValueError("Tipo de colocación de Masa Crítica no válido. Debe ser 'central', 'edge' o 'random'.")


    estrategias = inicializar_estrategias(red, nodos_masa_critica)
    fraccion_cooperacion_historia = [] # Lista para guardar la fracción de cooperadores en cada ronda

    for _ in range(rondas):
        recompensas = {nodo: 0 for nodo in red.nodes()} # Reiniciar recompensas para cada ronda

        # Jugar el juego y calcular las recompensas para cada nodo
        for nodo in red.nodes():
            for vecino in red.neighbors(nodo):
                recompensas[nodo] += calcular_recompensa(estrategias[nodo], estrategias[vecino], tipo_juego)

        # Actualizar las estrategias para la siguiente ronda
        siguientes_estrategias = {}
        for nodo in red.nodes():
            siguientes_estrategias[nodo] = actualizar_estrategia(red, nodo, estrategias, recompensas, nivel_homofilia, K, tipo_juego)
        estrategias = siguientes_estrategias # Las estrategias actualizadas se convierten en las estrategias actuales para la siguiente ronda

        conteo_cooperadores = sum(1 for estrategia in estrategias.values() if estrategia == 'C')
        fraccion_cooperacion = conteo_cooperadores / num_nodos
        fraccion_cooperacion_historia.append(fraccion_cooperacion)

    return fraccion_cooperacion_historia, red, estrategias # Retornar también red y estrategias finales


def graficar_red(red, estrategias, tipo_colocacion_masa_critica, filename):
    """Grafica la red con colores representando las estrategias y guarda la imagen en formato SVG."""
    plt.style.use(['science', 'ieee']) # Aplicar el estilo de scienceplots e ieee
    plt.figure(figsize=(3, 3)) # Ajustar el tamaño de la figura

    pos = nx.spring_layout(red)  # Layout para la red

    nodos_cooperadores = [nodo for nodo, estrategia in estrategias.items() if estrategia == 'C']
    nodos_defectores = [nodo for nodo, estrategia in estrategias.items() if estrategia == 'D']

    # Dibujar nodos cooperadores en azul cielo y más grandes
    nx.draw_networkx_nodes(red, pos, nodelist=nodos_cooperadores, node_color='#9B9ACA', node_size=15, edgecolors='#9B9ACA', linewidths=0.1)
    # Dibujar nodos defectores en azul marino
    nx.draw_networkx_nodes(red, pos, nodelist=nodos_defectores, node_color='#DEDEEE', node_size=15, edgecolors='#DEDEEE', linewidths=0.1)
    nx.draw_networkx_edges(red, pos, edge_color="#DEDEEE", alpha=1) # Dibujar aristas

    plt.title(f'{tipo_colocacion_masa_critica}') # Título como en el artículo
    plt.axis('off') # Quitar ejes
    plt.savefig(filename, format="svg", dpi=300, bbox_inches='tight') # Guardar la figura como SVG con alta resolución
    plt.close() # Cerrar la figura para liberar memoria


if __name__ == "__main__":
    num_nodos = 300 # Reducir nodos para visualización de red más clara
    tamaño_masa_critica_inicial = 150 # Ajustar masa crítica para red más pequeña
    nivel_homofilia = 0.75 # Ajustar homofilia para mayor o menor influencia de la homofilia (0 a 1)
    K = 0.5 # Ajustar K para mayor o menor ruido en la adopción de estrategias
    tipo_juego = "PDG" # Cambiar a "SG" para Juego de la Montaña de Nieve
    rondas = 200 

    tipos_colocacion_masa_critica = {
        'Nodos centrales': 'central',
        'Nodos aleatorios': 'random',
        'Nodos periféricos': 'edge'        
    }
    resultados_colocacion = {}

    # **Crear la red *FUERA* del bucle, una sola vez**
    red_unica = crear_red_libre_escala(num_nodos, m=1)

    for tipo_colocacion_nombre, tipo_colocacion_codigo in tipos_colocacion_masa_critica.items():
        historia_cooperacion, red_final, estrategias_finales = ejecutar_simulacion(
            # **Usar *red_unica* en cada simulación**
            red_unica,
            tamaño_masa_critica_inicial, tipo_colocacion_codigo, nivel_homofilia, K, tipo_juego, rondas
        )
        resultados_colocacion[tipo_colocacion_nombre] = historia_cooperacion

        # Graficar la red y guardarla como imagen SVG
        filename = f"red_{tipo_colocacion_codigo}.svg" # Cambiar extensión a .svg
        graficar_red(red_final, estrategias_finales, tipo_colocacion_nombre, filename)
        print(f"Red graficada y guardada como '{filename}'")


    # Define los colores para cada línea
    colors = ['#9B9ACA', '#7976B8', '#201F3D']

    # Graficar la evolución de la cooperación (como antes)
    plt.style.use(['science', 'ieee'])
    plt.figure(figsize=(3, 3))
    for (tipo_colocacion_nombre, historia), color in zip(resultados_colocacion.items(), colors):
        plt.plot(historia, label=f'{tipo_colocacion_nombre}', color=color)

    plt.xlabel("Iteraciones")
    plt.ylabel("Fracción de Cooperadores")
    plt.title(f"Evolución de la Cooperación")
    plt.legend()
    plt.grid(True)
    plt.savefig("evolucion_cooperacion.svg", format="svg", dpi=300, bbox_inches='tight') # Guardar gráfica de evolución como SVG
    plt.show()
    print(f"({tipo_juego}, Homofilia={nivel_homofilia}, K={K})")  # Mostrar los parámetros utilizados

    #Mostrar las redes creadas
    #Mostrar las imágenes generadas
    for tipo_colocacion_nombre in tipos_colocacion_masa_critica.keys():
        filename = f"red_{tipos_colocacion_masa_critica[tipo_colocacion_nombre]}.svg"
        display(SVG(filename))
````

````python
import networkx as nx
import numpy as np
import random
from IPython.display import SVG, display
import matplotlib.pyplot as plt
import scienceplots  # Importar la librería scienceplots
import os # Necesario para crear directorios si no existen

# --- Funciones Auxiliares ---

def crear_red_libre_escala(n, m=1, seed=None):
    """
    Crea una red libre de escala utilizando el modelo Barabási-Albert.
    Permite especificar una semilla para reproducibilidad.
    """
    return nx.barabasi_albert_graph(n, m, seed=seed)

def inicializar_estrategias(red, nodos_cooperadores_iniciales, estrategia_cooperar='C', estrategia_defectar='D'):
    """Inicializa las estrategias de los nodos de la red."""
    estrategias = {}
    for nodo in red.nodes():
        if nodo in nodos_cooperadores_iniciales:
            estrategias[nodo] = estrategia_cooperar
        else:
            estrategias[nodo] = estrategia_defectar
    return estrategias

def calcular_recompensa(estrategia_jugador, estrategia_vecino, tipo_juego="PDG"):
    """Calcula la recompensa (payoff) para un jugador en una interacción con un vecino."""
    if tipo_juego == "PDG":
        matriz_recompensa = {('C', 'C'): (3, 3), ('C', 'D'): (0, 5), ('D', 'C'): (5, 0), ('D', 'D'): (1, 1)}
    elif tipo_juego == "SG":
        matriz_recompensa = {('C', 'C'): (3, 3), ('C', 'D'): (1, 5), ('D', 'C'): (5, 1), ('D', 'D'): (0, 0)}
    else:
        raise ValueError("Tipo de juego no válido. Debe ser 'PDG' o 'SG'.")
    return matriz_recompensa[(estrategia_jugador, estrategia_vecino)][0]

def actualizar_estrategia(red, nodo, estrategias, recompensas, nivel_homofilia, K=0.1, tipo_juego="PDG"):
    """Actualiza la estrategia de un nodo utilizando la regla de Fermi, con influencia de la homofilia."""
    if not list(red.neighbors(nodo)):
        return estrategias[nodo]

    estrategia_actual = estrategias[nodo]
    recompensa_actual = recompensas[nodo]

    vecino = random.choice(list(red.neighbors(nodo)))
    estrategia_vecino = estrategias[vecino]
    recompensa_vecino = recompensas[vecino]

    if estrategia_actual == estrategia_vecino:
        factor_homofilia = 1 - nivel_homofilia
    else:
        factor_homofilia = 1

    try:
        if K <= 1e-9:
             probabilidad_adoptar = 1.0 if recompensa_vecino > recompensa_actual else 0.0
        else:
            delta_recompensa_scaled = np.clip((recompensa_actual - recompensa_vecino) / K, -700, 700)
            probabilidad_adoptar = (1 / (1 + np.exp(delta_recompensa_scaled))) * factor_homofilia
    except (OverflowError, FloatingPointError):
        probabilidad_adoptar = 0.0 if (recompensa_actual - recompensa_vecino) > 0 else 1.0


    if random.random() < probabilidad_adoptar:
        return estrategia_vecino
    else:
        return estrategia_actual

def ejecutar_simulacion_interna(red, tamaño_masa_critica_inicial, tipo_colocacion_masa_critica, nivel_homofilia, K, tipo_juego, rondas=200):
    """Función interna para ejecutar una simulación específica."""
    num_nodos = len(red.nodes())
    try:
        centralidad = nx.eigenvector_centrality(red, max_iter=1000, tol=1.0e-6)
    except nx.PowerIterationFailedConvergence:
        print("Advertencia: La centralidad de vector propio no convergió. Usando centralidad de grado en su lugar.")
        centralidad = nx.degree_centrality(red)

    if tipo_colocacion_masa_critica == 'central':
        nodos_masa_critica = sorted(centralidad, key=centralidad.get, reverse=True)[:tamaño_masa_critica_inicial]
    elif tipo_colocacion_masa_critica == 'edge':
        nodos_masa_critica = sorted(centralidad, key=centralidad.get)[:tamaño_masa_critica_inicial]
    elif tipo_colocacion_masa_critica == 'random':
        nodos_masa_critica = random.sample(list(red.nodes()), tamaño_masa_critica_inicial)
    else:
        raise ValueError("Tipo de colocación de Masa Crítica no válido.")

    estrategias = inicializar_estrategias(red, nodos_masa_critica)
    fraccion_cooperacion_historia = []

    for r in range(rondas):
        recompensas = {nodo: 0 for nodo in red.nodes()}
        for nodo in red.nodes():
            if list(red.neighbors(nodo)):
                for vecino in red.neighbors(nodo):
                    recompensas[nodo] += calcular_recompensa(estrategias[nodo], estrategias[vecino], tipo_juego)
            else:
                 recompensas[nodo] = 0

        siguientes_estrategias = {}
        for nodo in red.nodes():
            siguientes_estrategias[nodo] = actualizar_estrategia(red, nodo, estrategias, recompensas, nivel_homofilia, K, tipo_juego)
        estrategias = siguientes_estrategias

        conteo_cooperadores = sum(1 for est in estrategias.values() if est == 'C')
        fraccion_cooperacion = conteo_cooperadores / num_nodos if num_nodos > 0 else 0
        fraccion_cooperacion_historia.append(fraccion_cooperacion)

    return fraccion_cooperacion_historia, estrategias

# --- Función graficar_red ---
def graficar_red(red, estrategias, tipo_colocacion_masa_critica_nombre, filename):
    """Grafica la red con colores representando las estrategias y guarda la imagen en formato SVG."""
    plt.style.use(['science', 'ieee']) # Aplicar el estilo de scienceplots e ieee
    plt.figure(figsize=(3, 3)) # Ajustar el tamaño de la figura

    pos = nx.spring_layout(red, seed=42)  # Layout para la red (usar seed para reproducibilidad)

    nodos_cooperadores = [nodo for nodo, estrategia in estrategias.items() if estrategia == 'C']
    nodos_defectores = [nodo for nodo, estrategia in estrategias.items() if estrategia == 'D']

    # Dibujar nodos cooperadores en azul cielo y más grandes
    nx.draw_networkx_nodes(red, pos, nodelist=nodos_cooperadores, node_color='#9B9ACA', node_size=15, edgecolors='#9B9ACA', linewidths=0.1)
    # Dibujar nodos defectores en azul marino
    nx.draw_networkx_nodes(red, pos, nodelist=nodos_defectores, node_color='#DEDEEE', node_size=15, edgecolors='#DEDEEE', linewidths=0.1)
    nx.draw_networkx_edges(red, pos, edge_color="#DEDEEE", alpha=1) # Dibujar aristas

    plt.title(f'{tipo_colocacion_masa_critica_nombre}') # Título con el nombre del tipo de colocación
    plt.axis('off') # Quitar ejes
    plt.savefig(filename, format="svg", dpi=300, bbox_inches='tight') # Guardar la figura como SVG con alta resolución
    plt.close() # Cerrar la figura para liberar memoria


# --- Función Principal Empaquetada ---
def simular_y_graficar_cooperacion(
    num_nodos=300,
    m_barabasi=1,
    tamaño_masa_critica_inicial=150,
    nivel_homofilia=0.75,
    K=0.5,
    tipo_juego="PDG",
    rondas=200,
    output_dir="resultados_simulacion",
    network_seed=None,
    mostrar_en_notebook=True
):
    """
    Ejecuta la simulación completa de cooperación para diferentes colocaciones
    de masa crítica en la misma red base y genera las gráficas resultantes.
    Permite especificar una semilla para la generación de la red.
    """

    os.makedirs(output_dir, exist_ok=True)

    tipos_colocacion_masa_critica = {
        'Nodos centrales': 'central',
        'Nodos aleatorios': 'random',
        'Nodos periféricos': 'edge'
    }
    resultados_colocacion = {}
    archivos_red = {}

    print(f"Creando red base con {num_nodos} nodos (m={m_barabasi}, seed={network_seed})...")
    red_unica = crear_red_libre_escala(num_nodos, m=m_barabasi, seed=network_seed)
    print("Red base creada.")

    print(f"\nEjecutando simulaciones para el juego: {tipo_juego}...")
    for tipo_colocacion_nombre, tipo_colocacion_codigo in tipos_colocacion_masa_critica.items():
        print(f"  Simulando para: {tipo_colocacion_nombre}...")
        historia_cooperacion, estrategias_finales = ejecutar_simulacion_interna(
            red_unica,
            tamaño_masa_critica_inicial,
            tipo_colocacion_codigo,
            nivel_homofilia,
            K,
            tipo_juego,
            rondas
        )
        resultados_colocacion[tipo_colocacion_nombre] = historia_cooperacion

        filename_base = f"red_{tipo_colocacion_codigo}.svg"
        filename_path = os.path.join(output_dir, filename_base)
        # Llamar a la función graficar_red actualizada
        graficar_red(red_unica, estrategias_finales, tipo_colocacion_nombre, filename_path)
        archivos_red[tipo_colocacion_nombre] = filename_path
        print(f"    Red graficada y guardada como '{filename_path}'")

    print("\nGenerando gráfica de evolución...")
    colors = ['#9B9ACA', '#7976B8', '#201F3D']
    plt.style.use(['science', 'ieee'])
    plt.figure(figsize=(3, 3))
    for (tipo_colocacion_nombre, historia), color in zip(resultados_colocacion.items(), colors):
        plt.plot(historia, label=f'{tipo_colocacion_nombre}', color=color)

    plt.xlabel("Iteraciones")
    plt.ylabel("Fracción de Cooperadores")
    plt.title(f"Evolución Cooperación ({tipo_juego})")
    plt.legend()
    plt.grid(True)
    filename_evolucion = os.path.join(output_dir, f"evolucion_cooperacion_{tipo_juego.lower()}.svg")
    plt.savefig(filename_evolucion, format="svg", dpi=300, bbox_inches='tight')
    plt.close()
    print(f"Gráfica de evolución guardada como '{filename_evolucion}'")

    print(f"\nParámetros Utilizados: ({tipo_juego}, Homofilia={nivel_homofilia}, K={K}, Nodos={num_nodos}, MC={tamaño_masa_critica_inicial}, Rondas={rondas}, Seed Red={network_seed})")

    if mostrar_en_notebook:
        print("\nMostrando redes generadas (si está en un notebook):")
        try:
            for tipo_nombre, filepath in archivos_red.items():
                print(f"--- {tipo_nombre} ({tipo_juego}) ---")
                display(SVG(filepath))
            print(f"--- Evolución Cooperación ({tipo_juego}) ---")
            display(SVG(filename_evolucion))
        except NameError:
             print("  (No se pudo usar 'display'. Asegúrate de ejecutar en un entorno como Jupyter Notebook/Lab).")
        except FileNotFoundError as e:
             print(f"  Error al mostrar archivo: {e}. Asegúrate de que los archivos se generaron correctamente.")

    return resultados_colocacion

# --- Ejemplo de Uso ---
if __name__ == "__main__":
    print("Iniciando simulación de cooperación...")

    # --- Parámetros Comunes ---
    NUM_NODOS = 300 # Número de nodos en la red
    M_BARABASI = 1 # Parámetro m del modelo Barabási-Albert
    TAM_MC = 150 # Tamaño de la Masa Crítica inicial
    HOMOFILIA = 0.75 # Nivel de homofilia (0 a 1)
    K_RUIDO = 0.5 # Parámetro K para ruido en la adopción de estrategias
    NUM_RONDAS = 200 # Número de rondas de la simulación
    SEED_RED = 42 # Semilla para la generación de la red ("None" para aleatorio)
    MOSTRAR_NOTEBOOK = True # Cambiar a False si no estás en Jupyter

    # --- Simulación para Dilema del Prisionero (PDG) ---
    print(f"\n--- Ejecutando simulación para Dilema del Prisionero (PDG) ---")
    resultados_pdg = simular_y_graficar_cooperacion(
        num_nodos=NUM_NODOS,
        m_barabasi=M_BARABASI,
        tamaño_masa_critica_inicial=TAM_MC,
        nivel_homofilia=HOMOFILIA,
        K=K_RUIDO,
        tipo_juego="PDG",
        rondas=NUM_RONDAS,
        output_dir="Resultados_PDG",
        network_seed=SEED_RED,
        mostrar_en_notebook=MOSTRAR_NOTEBOOK
    )
    print("\nSimulación PDG completada.")

    # --- Simulación para Juego Montaña de Nieve (SG) ---
    print(f"\n--- Ejecutando simulación para Juego Montaña de Nieve (SG) ---")
    resultados_sg = simular_y_graficar_cooperacion(
        num_nodos=NUM_NODOS,
        m_barabasi=M_BARABASI,
        tamaño_masa_critica_inicial=TAM_MC,
        nivel_homofilia=HOMOFILIA,
        K=K_RUIDO,
        tipo_juego="SG",
        rondas=NUM_RONDAS,
        output_dir="Resultados_SG",
        network_seed=SEED_RED,
        mostrar_en_notebook=MOSTRAR_NOTEBOOK
    )
    print("\nSimulación SG completada.")

    print("\nAmbas simulaciones completadas usando la misma estructura de red inicial.")
````

````python
import networkx as nx
import numpy as np
import random
from IPython.display import SVG, display
import matplotlib.pyplot as plt
import scienceplots  # Importar la librería scienceplots
import os # Necesario para crear directorios si no existen
import re # Para sanitizar nombres de archivo

# --- Funciones Auxiliares (Sin cambios significativos aquí) ---

def crear_red_libre_escala(n, m=1, seed=None):
    """
    Crea una red libre de escala utilizando el modelo Barabási-Albert.
    Permite especificar una semilla para reproducibilidad.
    """
    return nx.barabasi_albert_graph(n, m, seed=seed)

def inicializar_estrategias(red, nodos_cooperadores_iniciales, estrategia_cooperar='C', estrategia_defectar='D'):
    """Inicializa las estrategias de los nodos de la red."""
    estrategias = {}
    for nodo in red.nodes():
        if nodo in nodos_cooperadores_iniciales:
            estrategias[nodo] = estrategia_cooperar
        else:
            estrategias[nodo] = estrategia_defectar
    return estrategias

def calcular_recompensa(estrategia_jugador, estrategia_vecino, tipo_juego="PDG"):
    """Calcula la recompensa (payoff) para un jugador en una interacción con un vecino."""
    # La lógica interna sigue usando las siglas "PDG" y "SG"
    if tipo_juego == "PDG":
        matriz_recompensa = {('C', 'C'): (3, 3), ('C', 'D'): (0, 5), ('D', 'C'): (5, 0), ('D', 'D'): (1, 1)}
    elif tipo_juego == "SG":
        matriz_recompensa = {('C', 'C'): (3, 3), ('C', 'D'): (1, 5), ('D', 'C'): (5, 1), ('D', 'D'): (0, 0)}
    else:
        raise ValueError("Tipo de juego no válido. Debe ser 'PDG' o 'SG'.")
    return matriz_recompensa[(estrategia_jugador, estrategia_vecino)][0]

def actualizar_estrategia(red, nodo, estrategias, recompensas, nivel_homofilia, K=0.1, tipo_juego="PDG"):
    """Actualiza la estrategia de un nodo utilizando la regla de Fermi, con influencia de la homofilia."""
    if not list(red.neighbors(nodo)):
        return estrategias[nodo]

    estrategia_actual = estrategias[nodo]
    recompensa_actual = recompensas[nodo]

    vecino = random.choice(list(red.neighbors(nodo)))
    estrategia_vecino = estrategias[vecino]
    recompensa_vecino = recompensas[vecino]

    if estrategia_actual == estrategia_vecino:
        factor_homofilia = 1 - nivel_homofilia
    else:
        factor_homofilia = 1

    try:
        if K <= 1e-9:
             probabilidad_adoptar = 1.0 if recompensa_vecino > recompensa_actual else 0.0
        else:
            delta_recompensa_scaled = np.clip((recompensa_actual - recompensa_vecino) / K, -700, 700)
            probabilidad_adoptar = (1 / (1 + np.exp(delta_recompensa_scaled))) * factor_homofilia
    except (OverflowError, FloatingPointError):
        probabilidad_adoptar = 0.0 if (recompensa_actual - recompensa_vecino) > 0 else 1.0

    if random.random() < probabilidad_adoptar:
        return estrategia_vecino
    else:
        return estrategia_actual

def ejecutar_simulacion_interna(red, tamaño_masa_critica_inicial, tipo_colocacion_masa_critica, nivel_homofilia, K, tipo_juego, rondas=200):
    """Función interna para ejecutar una simulación específica."""
    num_nodos = len(red.nodes())
    try:
        # Usar un valor de tol más pequeño si es necesario y posible
        centralidad = nx.eigenvector_centrality(red, max_iter=1000, tol=1.0e-8)
    except nx.PowerIterationFailedConvergence:
        print("Advertencia: La centralidad de vector propio no convergió. Usando centralidad de grado.")
        centralidad = nx.degree_centrality(red)
        # Asegurarse de que la centralidad de grado esté normalizada si es necesario
        # (degree_centrality ya lo devuelve normalizado por defecto)


    if tipo_colocacion_masa_critica == 'central':
        nodos_masa_critica = sorted(centralidad, key=centralidad.get, reverse=True)[:tamaño_masa_critica_inicial]
    elif tipo_colocacion_masa_critica == 'edge':
        nodos_masa_critica = sorted(centralidad, key=centralidad.get)[:tamaño_masa_critica_inicial]
    elif tipo_colocacion_masa_critica == 'random':
        # Usar random.seed si se necesita reproducibilidad en la selección aleatoria
        # random.seed(network_seed) # Podría pasarse como argumento si se requiere
        nodos_masa_critica = random.sample(list(red.nodes()), tamaño_masa_critica_inicial)
    else:
        raise ValueError("Tipo de colocación de Masa Crítica no válido.")

    estrategias = inicializar_estrategias(red, nodos_masa_critica)
    fraccion_cooperacion_historia = []

    for r in range(rondas):
        recompensas = {nodo: 0 for nodo in red.nodes()}
        for nodo in red.nodes():
            vecinos_nodo = list(red.neighbors(nodo)) # Obtener lista de vecinos una vez
            if vecinos_nodo:
                for vecino in vecinos_nodo:
                    # Asegurar que la estrategia del vecino exista (aunque debería con red conectada)
                    if vecino in estrategias:
                        recompensas[nodo] += calcular_recompensa(estrategias[nodo], estrategias[vecino], tipo_juego)
                    else:
                        # Manejo de caso improbable
                        print(f"Advertencia: Vecino {vecino} del nodo {nodo} no encontrado en estrategias.")
            else:
                 recompensas[nodo] = 0

        siguientes_estrategias = {}
        for nodo in red.nodes():
            siguientes_estrategias[nodo] = actualizar_estrategia(red, nodo, estrategias, recompensas, nivel_homofilia, K, tipo_juego)
        estrategias = siguientes_estrategias

        conteo_cooperadores = sum(1 for est in estrategias.values() if est == 'C')
        fraccion_cooperacion = conteo_cooperadores / num_nodos if num_nodos > 0 else 0
        fraccion_cooperacion_historia.append(fraccion_cooperacion)

    return fraccion_cooperacion_historia, estrategias

def graficar_red(red, estrategias, tipo_colocacion_masa_critica_nombre, filename):
    """Grafica la red con colores representando las estrategias y guarda la imagen en formato SVG."""
    plt.style.use(['science', 'ieee'])
    plt.figure(figsize=(3, 3))
    pos = nx.spring_layout(red, seed=42)

    nodos_cooperadores = [nodo for nodo, estrategia in estrategias.items() if estrategia == 'C']
    nodos_defectores = [nodo for nodo, estrategia in estrategias.items() if estrategia == 'D']

    nx.draw_networkx_nodes(red, pos, nodelist=nodos_cooperadores, node_color='#9B9ACA', node_size=15, edgecolors='#9B9ACA', linewidths=0.1)
    nx.draw_networkx_nodes(red, pos, nodelist=nodos_defectores, node_color='#DEDEEE', node_size=15, edgecolors='#DEDEEE', linewidths=0.1)
    nx.draw_networkx_edges(red, pos, edge_color="#DEDEEE", alpha=1)

    plt.title(f'{tipo_colocacion_masa_critica_nombre}')
    plt.axis('off')
    plt.savefig(filename, format="svg", dpi=300, bbox_inches='tight')
    plt.close()

# --- Función Principal Empaquetada (Modificada) ---

def simular_y_graficar_cooperacion(
    num_nodos=300,
    m_barabasi=1,
    tamaño_masa_critica_inicial=150,
    nivel_homofilia=0.75,
    K=0.5,
    tipo_juego_sigla="PDG", # Usar la sigla internamente
    rondas=200,
    output_dir_base="resultados_simulacion", # Directorio base
    network_seed=None,
    mostrar_en_notebook=True
):
    """
    Ejecuta la simulación completa de cooperación para diferentes colocaciones
    de masa crítica en la misma red base y genera las gráficas resultantes.
    Permite especificar una semilla para la generación de la red.
    Usa nombres completos de juegos para directorios y títulos.
    """
    # Mapeo de siglas a nombres completos y nombres para archivos/directorios
    mapa_juegos = {
        "PDG": {"nombre_completo": "Dilema del prisionero", "nombre_archivo": "Dilema_Prisionero"},
        "SG": {"nombre_completo": "Dilema de la nieve acumulada", "nombre_archivo": "Nieve_Acumulada"}
    }

    if tipo_juego_sigla not in mapa_juegos:
        raise ValueError("Sigla de tipo de juego no válida. Debe ser 'PDG' o 'SG'.")

    juego_info = mapa_juegos[tipo_juego_sigla]
    nombre_juego_completo = juego_info["nombre_completo"]
    nombre_juego_archivo = juego_info["nombre_archivo"]

    # Crear directorio de salida específico para el juego
    output_dir = os.path.join(output_dir_base, nombre_juego_archivo) # Usar nombre de archivo para directorio
    os.makedirs(output_dir, exist_ok=True)

    tipos_colocacion_masa_critica = {
        'Nodos centrales': 'central',
        'Nodos aleatorios': 'random',
        'Nodos periféricos': 'edge'
    }
    resultados_colocacion = {}
    archivos_red = {}

    print(f"Creando red base con {num_nodos} nodos (m={m_barabasi}, seed={network_seed})...")
    red_unica = crear_red_libre_escala(num_nodos, m=m_barabasi, seed=network_seed)
    print("Red base creada.")

    # Usar el nombre completo del juego en los mensajes
    print(f"\nEjecutando simulaciones para: {nombre_juego_completo}...")
    for tipo_colocacion_nombre, tipo_colocacion_codigo in tipos_colocacion_masa_critica.items():
        print(f"  Simulando para: {tipo_colocacion_nombre}...")
        historia_cooperacion, estrategias_finales = ejecutar_simulacion_interna(
            red_unica,
            tamaño_masa_critica_inicial,
            tipo_colocacion_codigo,
            nivel_homofilia,
            K,
            tipo_juego_sigla, # Pasar la sigla a la función interna
            rondas
        )
        resultados_colocacion[tipo_colocacion_nombre] = historia_cooperacion

        # Usar el código de colocación para el nombre base del archivo de red
        filename_base = f"red_{tipo_colocacion_codigo}.svg"
        filename_path = os.path.join(output_dir, filename_base)
        graficar_red(red_unica, estrategias_finales, tipo_colocacion_nombre, filename_path)
        archivos_red[tipo_colocacion_nombre] = filename_path
        print(f"    Red graficada y guardada como '{filename_path}'")

    print("\nGenerando gráfica de evolución...")
    colors = ['#9B9ACA', '#7976B8', '#201F3D']
    plt.style.use(['science', 'ieee'])
    plt.figure(figsize=(3, 3))
    for (tipo_colocacion_nombre, historia), color in zip(resultados_colocacion.items(), colors):
        plt.plot(historia, label=f'{tipo_colocacion_nombre}', color=color)

    plt.xlabel("Iteraciones")
    plt.ylabel("Fracción de Cooperadores")
    # Usar el nombre completo del juego en el título de la gráfica
    plt.title(f"Evolución Cooperación\n({nombre_juego_completo})") # Salto de línea para mejor formato
    plt.legend(fontsize='small') # Ajustar tamaño de leyenda si es necesario
    plt.grid(True)
    # Usar nombre de archivo sanitizado para la gráfica de evolución
    filename_evolucion_base = f"evolucion_cooperacion_{nombre_juego_archivo.lower()}.svg"
    filename_evolucion = os.path.join(output_dir, filename_evolucion_base)
    plt.savefig(filename_evolucion, format="svg", dpi=300, bbox_inches='tight')
    plt.close()
    print(f"Gráfica de evolución guardada como '{filename_evolucion}'")

    # Usar el nombre completo del juego en el mensaje final
    print(f"\nParámetros Utilizados: ({nombre_juego_completo}, Homofilia={nivel_homofilia}, K={K}, Nodos={num_nodos}, MC={tamaño_masa_critica_inicial}, Rondas={rondas}, Seed Red={network_seed})")

    if mostrar_en_notebook:
        print("\nMostrando resultados generados (si está en un notebook):")
        try:
            # Mostrar redes
            for tipo_nombre, filepath in archivos_red.items():
                print(f"--- {tipo_nombre} ({nombre_juego_completo}) ---")
                if os.path.exists(filepath):
                    display(SVG(filepath))
                else:
                    print(f"  Archivo no encontrado: {filepath}")
            # Mostrar gráfica de evolución
            print(f"--- Evolución Cooperación ({nombre_juego_completo}) ---")
            if os.path.exists(filename_evolucion):
                display(SVG(filename_evolucion))
            else:
                 print(f"  Archivo no encontrado: {filename_evolucion}")

        except NameError:
             print("  (No se pudo usar 'display'. Asegúrate de ejecutar en un entorno como Jupyter Notebook/Lab).")
        except Exception as e: # Capturar otras posibles excepciones
             print(f"  Ocurrió un error al intentar mostrar los archivos: {e}")


    return resultados_colocacion

# --- Ejemplo de Uso Modificado ---
if __name__ == "__main__":
    print("Iniciando simulación de cooperación...")

    # --- Parámetros Comunes ---
    NUM_NODOS = 300
    M_BARABASI = 1
    TAM_MC = 150
    HOMOFILIA = 0.8
    K_RUIDO = 0.5
    NUM_RONDAS = 200
    SEED_RED = 42
    MOSTRAR_NOTEBOOK = True # Cambiar a False si no estás en Jupyter
    DIRECTORIO_BASE = "Resultados_Simulaciones" # Directorio base para todos los resultados

    # --- Simulación para Dilema del prisionero (PDG) ---
    print(f"\n--- Ejecutando simulación para Dilema del prisionero ---")
    resultados_pdg = simular_y_graficar_cooperacion(
        num_nodos=NUM_NODOS,
        m_barabasi=M_BARABASI,
        tamaño_masa_critica_inicial=TAM_MC,
        nivel_homofilia=HOMOFILIA,
        K=K_RUIDO,
        tipo_juego_sigla="PDG", # <--- Pasar la SIGLA aquí
        rondas=NUM_RONDAS,
        output_dir_base=DIRECTORIO_BASE, # <--- Pasar directorio BASE
        network_seed=SEED_RED,
        mostrar_en_notebook=MOSTRAR_NOTEBOOK
    )
    print("\nSimulación Dilema del prisionero completada.")

    # --- Simulación para Dilema de la nieve acumulada (SG) ---
    print(f"\n--- Ejecutando simulación para Dilema de la nieve acumulada ---")
    resultados_sg = simular_y_graficar_cooperacion(
        num_nodos=NUM_NODOS,
        m_barabasi=M_BARABASI,
        tamaño_masa_critica_inicial=TAM_MC,
        nivel_homofilia=HOMOFILIA,
        K=K_RUIDO,
        tipo_juego_sigla="SG", # <--- Pasar la SIGLA aquí
        rondas=NUM_RONDAS,
        output_dir_base=DIRECTORIO_BASE, # <--- Pasar directorio BASE
        network_seed=SEED_RED, # Mismo SEED
        mostrar_en_notebook=MOSTRAR_NOTEBOOK
    )
    print("\nSimulación Dilema de la nieve acumulada completada.")

    print("\nAmbas simulaciones completadas usando la misma estructura de red inicial.")
    print(f"Resultados guardados en el directorio base: '{DIRECTORIO_BASE}'")
````
