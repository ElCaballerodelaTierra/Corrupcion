Se crea una función que mide la distribución de grado, el coeficiente de agrupamiento, la propiedad de mundo pequeño y la asortativdad en una red. 
Esto con el fin de conocer si una red tiene las propiedades de de una red corrupta.

```python
import networkx as nx
import matplotlib.pyplot as plt

def analizar_red(G):
    # Calcular el coeficiente de asortatividad de grado
    assortativity_coeff = nx.degree_assortativity_coefficient(G)
    
    # Calcular el coeficiente de agrupamiento global
    clustering_coeff = nx.average_clustering(G)
    
    # Calcular el camino medio más corto
    try:
        average_path_length = nx.average_shortest_path_length(G)
    except nx.NetworkXError:
        average_path_length = float('inf')
    
    # Obtener los grados de todos los nodos
    degrees = [G.degree(n) for n in G.nodes()]
    
    # Visualizar la distribución de grados
    plt.hist(degrees, bins=30)
    plt.xlabel('Grado')
    plt.ylabel('Frecuencia')
    plt.title('Distribución de Grado')
    plt.show()
    
    # Imprimir los resultados
    print(f"Coeficiente de Asortatividad: {assortativity_coeff}")
    print(f"Coeficiente de Agrupamiento: {clustering_coeff}")
    print(f"Camino Medio Más Corto: {average_path_length}")

# Ejemplo de uso con una red libre de escala
G = nx.barabasi_albert_graph(1000, 5)
analizar_red(G)

# Ejemplo de uso con una red de mundo pequeño
G_ws = nx.watts_strogatz_graph(1000, 6, 0.04)
analizar_red(G_ws)
```

Se añade el cálculo de la pendiente y su visualización en el gráfico

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize

def log_likelihood(params, x, y):
    a, b = params
    predicted_y = a * x + b
    log_likelihood_value = -np.sum((y - predicted_y)**2)
    return -log_likelihood_value

def ajustar_linea_mle(x, y):
    initial_params = [1, 0]
    result = minimize(log_likelihood, initial_params, args=(x, y), method='L-BFGS-B')
    return result.x

def analizar_red(G):
    # Calcular el coeficiente de asortatividad de grado
    assortativity_coeff = nx.degree_assortativity_coefficient(G)
    
    # Calcular el coeficiente de agrupamiento global
    clustering_coeff = nx.average_clustering(G)
    
    # Calcular el camino medio más corto
    try:
        average_path_length = nx.average_shortest_path_length(G)
    except nx.NetworkXError:
        average_path_length = float('inf')
    
    # Obtener los grados de todos los nodos
    degrees = [G.degree(n) for n in G.nodes()]
    
    # Visualizar la distribución de grados en escala logarítmica solo en el eje y
    plt.hist(degrees, bins=30, log=True, edgecolor='black', alpha=0.7)
    plt.xlabel('Grado')
    plt.ylabel('Frecuencia (escala logarítmica)')
    plt.title('Distribución de Grado')
    plt.yscale('log')
    
    # Convertir grados a un histograma
    hist, bin_edges = np.histogram(degrees, bins=30)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

    # Filtrar valores mayores que 0 para evitar problemas con el logaritmo
    mask = hist > 0
    hist = hist[mask]
    bin_centers = bin_centers[mask]

    # Ajustar la línea recta usando MLE en el log-log
    log_bin_centers = np.log10(bin_centers)
    log_hist = np.log10(hist)
    a, b = ajustar_linea_mle(log_bin_centers, log_hist)

    # Dibujar la línea de tendencia recta
    x_fit = np.linspace(min(bin_centers), max(bin_centers), 100)
    y_fit = 10**(a * np.log10(x_fit) + b)
    plt.plot(x_fit, y_fit, linestyle='--', color='red', label=f'Tendencia (MLE): y = {a:.2f}x + {b:.2f}')
    plt.legend()
    
    plt.show()
    
    # Imprimir los resultados
    print(f"Coeficiente de Asortatividad: {assortativity_coeff}")
    print(f"Coeficiente de Agrupamiento: {clustering_coeff}")
    print(f"Camino Medio Más Corto: {average_path_length}")

# Ejemplo de uso con una red libre de escala
G = nx.barabasi_albert_graph(1000, 5)
analizar_red(G)

# Ejemplo de uso con una red de mundo pequeño
G_ws = nx.watts_strogatz_graph(1000, 6, 0.04)
analizar_red(G_ws)
```

Otro enfoque mediante la construcción del método de máxima verosimulitud y después aplicándolo al gráfico de ditribución de grados de los nodos

Ajustando a una Ley de Potencias

Para determinar si la distribución de grados de tu red sigue una ley de potencias, necesitarás realizar un ajuste lineal en una escala logarítmica doble. Esto se debe a que una ley de potencias se expresa como:

P(k) ~ k^(-gamma)

Donde:

    P(k) es la probabilidad de que un nodo tenga un grado k.
    gamma es el exponente de la ley de potencias (también conocido como índice de escala).

Tomando el logaritmo de ambos lados de la ecuación, obtenemos:

log(P(k)) ~ -gamma * log(k)

Esto significa que si graficamos log(P(k)) contra log(k), esperamos obtener una línea recta con pendiente -gamma

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Crear una red (ejemplo: Barabási-Albert)
G = nx.barabasi_albert_graph(1000, 2)

# Obtener los grados de los nodos y calcular la distribución de frecuencia
degrees = list(dict(nx.degree(G)).values())
unique_degrees, counts = np.unique(degrees, return_counts=True)
probs = counts / len(G)

# Eliminar grados con frecuencia cero
non_zero_probs = probs[probs > 0]
non_zero_degrees = unique_degrees[probs > 0]

# Tomar el logaritmo de las probabilidades y los grados
log_probs = np.log(non_zero_probs)
log_degrees = np.log(non_zero_degrees)

# Realizar un ajuste lineal
slope, intercept, r_value, p_value, std_err = linregress(log_degrees, log_probs)

# Calcular el exponente de la ley de potencias
gamma = -slope

# Crear el gráfico
plt.loglog(unique_degrees, probs, 'o', label='Distribución de Grados')
plt.plot(unique_degrees, np.exp(intercept + slope * np.log(unique_degrees)), '--', label='Ajuste de Ley de Potencias')
plt.xlabel('Grado (log)')
plt.ylabel('Probabilidad (log)')
plt.legend()
plt.title('Distribución de Grados y Ajuste de Ley de Potencias')
plt.show()

# Graficar con escala logarítmica únicamente en el eje y
plt.plot(non_zero_degrees, non_zero_probs, 'o-', label='Distribución de Grados')
plt.yscale('log')
plt.xlabel('Grado')
plt.ylabel('Probabilidad')
plt.title('Distribución de Grados en una Red')
plt.legend()
plt.show()

# Imprimir el exponente y el coeficiente de correlación
print("Exponente de la ley de potencias (gamma):", gamma)
print("Coeficiente de correlación (r):", r_value)
```
